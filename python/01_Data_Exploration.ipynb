```python
import pandas as pd

# Load dataset
df = pd.read_csv('../data/Food-Delivery-App.csv')

```

```python
# Check first 5 rows
print(df.head())
```

```python
# Check columns and data types
print(df.info())
```

```python
# Check basic stats for numeric columns
print(df.describe())
```

```python
# Count missing values per column
print(df.isnull().sum())
```

```python
import numpy as np

# Replace 'Not given' with NaN
df['rating'] = df['rating'].replace('Not given', np.nan)

# Convert rating to numeric
df['rating'] = pd.to_numeric(df['rating'])

# Check result
print(df['rating'].dtype)
print(df['rating'].isnull().sum())
```

```python
df['total_order_time'] = df['food_preparation_time'] + df['delivery_time']
```

```python
df['delivery_speed'] = df['delivery_time'].apply(
    lambda x: 'Fast' if x <= 30 else 'Slow'
)
```
